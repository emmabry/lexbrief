```python
gold_summaries = []

with open('/Users/emma/msc_project/data/eur-lexsum/raw-data/train.target', 'r') as f:
    for i in range(5):
        line = f.readline().strip()
        if line:
            gold_summaries.append(line)
```

```python
from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

total_rouge1 = 0.0
total_rouge2 = 0.0
total_rougeL = 0.0

for pred_sum, gold_sum in zip(GPTsummaries, gold_summaries):
    scores = scorer.score(gold_sum, pred_sum)
    print(f"Predicted Summary: {pred_sum}")
    print(f"Gold Summary: {gold_sum}")
    print(f"ROUGE-1: {scores['rouge1'].fmeasure:.4f}")
    total_rouge1 += scores['rouge1'].fmeasure
    total_rouge2 += scores['rouge2'].fmeasure
    total_rougeL += scores['rougeL'].fmeasure

n = len(GPTsummaries)
avg_rouge1 = total_rouge1 / n
avg_rouge2 = total_rouge2 / n
avg_rougeL = total_rougeL / n

print(f"\nAverage ROUGE-1 F1: {avg_rouge1:.4f}")
print(f"Average ROUGE-2 F1: {avg_rouge2:.4f}")
print(f"Average ROUGE-L F1: {avg_rougeL:.4f}")
```
